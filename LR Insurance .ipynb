Text columns convert to numeric using labelencoder from sklearn

```python
from sklearn.preprocessing import LabelEncoder #import the labelencoder
```

```python
sex_encode=LabelEncoder()
df['sex']=sex_encode.fit_transform(df['sex'])
smoker_encode=LabelEncoder()
df['smoker']=smoker_encode.fit_transform(df['smoker'])
region_encode=LabelEncoder()
df['region']=region_encode.fit_transform(df['region'])
```

```python
df.head()
```

#using standard scalar from sklearn becoz we have a large number in the total price and we have convert to the small number of information we get

```python
from sklearn.preprocessing import StandardScaler #import the standard scalar function to customize the large num to covert the small no of value 
```

```python
charges_scaler=StandardScaler()
df['charges']=charges_scaler.fit_transform(df[['charges']])
```

```python
df.head() #we can check the standardscaler function it is working or not
```

# Splitting the data x(feature,independent variable) and y(target,dependent variable)

```python
x=df.drop(['charges'],axis=1)
x.head()
```

```python
y=df[['charges']]
y.head()
```

# Train Test splitting 

```python
from sklearn.model_selection import train_test_split #import train test split func
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=55)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# now we can import the linear regression model

```python
from sklearn.linear_model import LinearRegression
```

now creating a empty linear reg model

```python
linear_model=LinearRegression()
linear_model
```

```python
# Training the data only x_train,y_train these data only and thenwe can training the model
```

```python
linear_model.fit(x_train,y_train)
```

```python
y_predict=linear_model.predict(x_test)
y_predict
```

# Evaluation the model

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_predict)   #R² Score: Values closer to 1 indicate a better fit.
r2
```

```python
from sklearn.metrics import mean_squared_error
```

```python
accuracy=1-mean_squared_error(y_test,y_predict)
accuracy
```
